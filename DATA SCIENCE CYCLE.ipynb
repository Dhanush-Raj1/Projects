# 3.DATA PREPARATION

# 3.1 Encoding Categorical Features  
  - Encoding categorical features is one of the important steps in Data Preparation.   
  - Most machine learning models can understand only numerical values therefore it is necessary to convert the categorical values into numerical values so that our model can understand and extract information from the features.
  - There are two types of features(variables) while encoding   
&nbsp;  
**Ordinal variable:**    
    Those variables have a rank or order in it. And those rank or order can be used to convert it into numerical features.  
    Eg: Ratings, Temperature(low,medium,hot,cold)  
**Nominal Variable**  
    Those variables does not have a rank or order, they are just normal categories.  
    Eg: City names, Address  
  
&nbsp;      
   - Cheatsheet    
   
&nbsp;
## Methods to Encode Categorical Features:
### 1. One Hot Encoder / Dummy Encoder (creates columns)
   - One Hot encoder converts the categorical values of a feature by creating new columns for each unique categorical values.   
   - It assigns 1 where a particular categorical value present and 0 if it is not.  
   - One Hot encoder increases the dimension or number of columns if we have more number of unique categorical values to encode. And as a result it leads to Curse of Dimensionality. 
   - For Regression model we can use drop_first parameter and for classification model is adviced to use all the columns.
   - This encoder is suitable for algorithms that looks at "ALL" the featurs simultaneously during training. Linear regression, support vector machines, neural networks, clustering algorithms works in such way.  
   <font color=red>from sklearn.preprocessing import OneHotEncoder  
    onehot=OneHotEncoder()  
    data['feature_encoded']=onehot.fit_transform(data['feature'])    
    data</font>       
   - Dummy Encoder performs similar to one hot encoder but it leads to Dummy variable trap.  
    <font color=red>import pandas as pd  
    data['feature_encoded']=pd.get_dummies(data['feature'],drop_first=True)  
    data</font>  
&ensp;    
**One Hot Encoder With Multiple Categories**  
   - Suppose we have 30 different categorical values in a feature and we only want to use one hot encoder but the result will lead to 30 new columns and this leads to curse of dimensionality and poor model performance. 
   - In such cases we can use only the top 10 repeating categories i.e 10 categories that have observed the most by this we can reduce the number of columns .  
      
&nbsp;  
### 2. Label Encoder   
   - Label Encoder converts the categorical values of a feature by using the rank or order .  
   - By using label encoder we cannot assign ranks manually the encoder by default assigns ranks as per the order of the   values which leads to mis-lead information.
   - Lets say we have temperate feature with values low, mediam, hot, cold. The rank of those values should by low=1,   median=2, hot=3, cold=4. But Label encoder assigns rank by just seeing the order of the values if hot is the first value then the encoder converts it as 1 and similarly if we have cold as second value then it is 2.  
    <font color=red>from sklearn.preprocessing import LabelEncoder  
    label=LabelEncoder()  
    data['feature_encoded']=label.fit_transform(data['feature'])  
    data</font>  
    
&nbsp;
### 3. Ordinal Encoder 
   - Ordinal Encoder is similar to label encoder as it uses the rank or order but the difference is we can assign the ranks manually to the values.  
    <font color=red>from category_encoders import OrdinalEncoder    
    ordinal=OrdinalEncoder()  
    order=[{'col': 'temperature', 'mapping': {"low": 1,  "medium": 2, "hot": 3, "cold":4}}]  
    data['feature_encoded']=ordinal.fit_transform(data['feature'],mapping=order)  
    data</font>  
 
&nbsp;
### 4. Frequency Encoder / Count Encoder
   - Frequency Encoder uses the frequency or the count of a categorical value of a feature to convert it into numerical feature.
   - One drawback while using count encoder is that if our dataset contains two different categorical values with the same frequency then it is hard to distinguish between them.    
    <font color=red>from category_encoders import CountEncoder  
    count=CountEncoder()  
    data['feature_encoded']=count.fit_transform(data['feature'])  
    data</font>  
    or  
    <font color=red>frequency=data.groupby('feature').size()/len(data)  
    data['Feature_encoded']=data['feature'].map(frequency)  
    data</font>      
    or      
   - By using **SUNBIRD** library    
   <font color=red>from sunbird.category_encoding import frequency_encoding    
    frequency_encoding(data,'feature')    
    data</font>    
  
&nbsp;
### 5. Sum Encoder  
   - Sum Encoder is similar to one hot encoder except it uses 0, 1, -1 to convert values by assignning new columns. 
   - Compares the mean of the dependent variable (target) for a given level of a categorical column to the overall mean of the target.  
   - Sum encoder is a contrast encoder.  
   <font color=red>from category_encoders import SumEncoder  
    sumencoder=SumEncoder()  
    data['feature_encoded']=sumencoder.fit_transform(data['feature'])  
    data</font>  

&nbsp;
### 6. Binary Encoder
   - Binary encoder is similar to one hot encoder it converts the categorical columns into binary integers.
   - For example: hot, very hot, cold binary encoder assigns binary values to those values like for hot=001, veryhot=010, cold=100 and each binary value forms seperate column and totally three columns with values for hot 0 in first column, 0 in second column, 1 in third columns.  
   - If there are n unique categories, then binary encoding results in the only log(base 2)ⁿ features.
   - The number of dimensions is lesser compared to One hot encoder.  
    <font color=red>from category_encoders import BinaryEncoder  
    binary=BinaryEncoder()  
    data['feature_encoded']=binary.fit_transform(data['feature'])
    data</font>  
      
&nbsp;  
### 7. Hashing Encoder / Feature Hashing  
   - Hash encoding is similar to one hot encoding it converts categorical values into new columns except we can assign the number of columns to be created.   
   - In other words we can control the output dimensions.  
   - With Hashing, the number of dimensions will be far less than the number of dimensions with encoding like One Hot Encoding. This method is advantageous when the cardinality of categorical is very high.  
   <font color=red>from category_encoders import HashingEncoder  
    hash=HashingEncoder()  
    data['feature_encoded']=hash.fit_transform(data['feature'],n_components=5)  
    data</font>  
      
&nbsp;  
### 8. Helmert Encoder  
   - Helmert encoder compares the mean of a particular value of a categorical column to other values of the column.  
   - Lets say we have values like warm, hot, veryhot, cold the encoder compares the mean of warm with hot, veryhot,cold and the mean of hot with veryhot,cold and the mean of veryhot with cold.    
   - This encoder is useful when there is a order in a categorical column.  
   - Helmert encoder is a contrast encoder.Contrast coding allows for recentering of categorical variables such that the intercept of a model is not the mean of one level of a category, but instead the mean of all data points in the data set.

# 3.2 Handling Missing Values  

  * Missing values or null value occure due to many reasons. But the point missing values have a negative impact of our model performance. Hence, its a must to handle missing values.  
  * Different types of missing values:  

### 1. Missing Completely at Random (MCAR)   
   - A variable is missing completely at random if the probability of being missing is same for all the observations. The data   missing is totally because of randomness.  
   - When data is MCAR there is absolutely no relationship between the missing values and any other values of the dataset.  
   - Missing values are completely independent of other data.    
   - Since there is no pattern for the missing values , we can predict those missing values with commom knowledge or idea , with assumptions , with domain knowledge .   
   - Here the assumption we can make does not relate with other data.  
   - For example: Men usually will not complete the entire survey they may skip some questions out of no reasons.  

###  2. Missing at Random (MAR) 
   - A variable missing at random means that there is systematic relationship betweent the missing values and the observed data
but not the unobserved data. The data missing is not because of randomness but there is reason or relationship with other data.
   - There is a relationship with the missing values and the observed data.  
   - Since there is a pattern for the missing values , we can predict those missing values based on other observed data.  
   - Here the assumption we can make solelly relate with other data .  
   - For example: Women usually will not reveal there age there is a specific reason and relationship.  

### 3. Missing Not at Random (MNAR)    
   - A variable is missing not at random if the probability of being missing is not the same for all the observations there is a systematic order or reason for those missing values.     
   - And there is a absolute specific relationship between the missing values and any other values of the dataset.  
   - We cannot make assumptions and predictions based on other data.  
   - For example: People may not reveal there income if their income is low .

# 3.3 Feature Scaling    
   - Feature Scaling is the process of scaling the features of a dataset in an equal range.    
   - Real life datasets often contain features that vary in magnitude, units and range and machine wont't consider and treat all those features in same scale. 
   - For eg: Income has high range of value compared to age and machine learning model won't treat those features equally as income has high influence compared to age over our target columns.   
   - What feature scaling does is it **keeps the distribution of the data, the correlation and covariance absolutely the same**    however scales every independent feature  to a smaller scale. 
   
   - Therefore it is necessary to scale the feature of our dataset in a similar range so that machine learning model can       interpret and treat all our features in a same scale.    
   - Algorithms where feature scaling matters: (any algorithm that uses the distance of features)  
      - K Nearest Neighbors    
      - Performing PCA  
      - K Means Clustering  
   - One of the main disadvantage of feature scaling is that it is **affected by outliers**, even a single outliers makes things messy. Treat outliers before performing scaling.  
          
&nbsp;  
### 1. Standardization / Z-Score Normalization  
   - Standardization is a process in which the values are scaled with Z-score.
   - Calculation of Z-score 
   - The outcome of standardization is it changes the mean of a feature into 0 (μ = 0) and standard deviation into 1 (σ =1).
   - This technique is best to use if the feature is normally distributed.  
   - It is affected by outliers.
   - It is useful for common regression tasks.  
   <font color=red>from sklearn.preprocessing import StandardScaler

# 3.5 Feature Selection  
   - Feature selection is a process of selecting only the most relevant features from original set of features by removing irrelevant or noisy features. 
   - Feature selection process selects only those feature that are important for a model to train and excludes irrelevant feature which will negatively impact our model.  
   - In building a machine learning model real life datasets contains lots of features among which some will be relevant for model building and some will not be. If we build the model with all the existing features it will affect the overall performance, accuracy of the model and it leads to **Curse Of Dimensionality**. Therefore it is necessary to perform feature selection.  
     
## Benefits of Feature selection  
   1. It reduces CURSE OF DIMENSIONALITY.  
   2. It leads to better overall performance and simplification of model.  
   3. It reduces training time of model.  
   4. It reduces overfitting by making a more generalized model.

   - Wrapper methods selects features by creating all possible subsets of features and asses their quality by evaluating their performance over target varaible by applying a machine learning algorithm(classifier).  ## 1.Forward Feature Selection
   - Forward selection is an iterative process which begins with an empty set of features in the model. After each iteration, it keeps adding features which best improves the performance of the model. This process continues until addition of a new feature does not improve the performance of the model.
   - FFS method will take time to work on large datasets therefore it is advised to use on smaller datasets
      
   
 ## 2.Backward Feature Elimination
   - Backward elimination is also an iterative process, but it is the opposite of Forward feature selection. The process begins with all the features in the model, after every iteration it eliminates/removes the feature that does not improve the performance of the model. This process continues until elimination of a feature does not improve the performance of the mode.  
   - BFE method will also take time to work on large datasets therefore it is advised to use on smaller datasets.

 ## 3.Exhaustive Feature Selection
      Exhaustive selection method is the most robust method. This is a brute-force selection method which tries each and every possible combination of feature set and returns the best performing subset.
   
 ## 4.Recursive Feature elimination
   - Recursive feature elimination (RFE) is a feature selection method that fits a model and removes the weakest feature until the specified number of features is reached.
   -  RFE selects features by considering smaller and smaller set of features.First, the estimator is trained on the initial set of features and the importance of each feature is obtained either through a coef_ attribute or through a feature_importances_ attribute. Then, the least important features are pruned from the current set of features. 
   - By recursively eliminating a small number of features per loop, RFE attempts to eliminate dependencies and collinearity that may exist in the model.

# Errors 
  - Error an important aspect of evalutating the performance of a model.
  - Error usually refers to the difference between the actual value(ground truth) and the predicted value.
  - The error for any supervised Machine Learning algorithm comprises of 3 parts:
     - Bias error
     - Variance error
     - The noise
  - The noise is the irreducible error, we cannot eliminate or reduce it while the other two are reducible.

## Bias    
  - Bias is defined as phenomenon that skews or affects the result of an algorithm in favor or against an idea.  
  - We can define bias as the error between average model prediction and the ground truth.    
  - When we talk about bias affecting the result of an model we usually refer to the result of the **training data not the test data.**  
  - **Low Bias** when our model has low bias it means that our model performs well on the training data and gives us good results on the train dataset. 
  - **High Bias** when our model has high bias it means that our model does not learn much from the training data and as a result the performance is poor.   
  - Mathematically, 
     - lets say our input variable as X and output variable as Y and we find the relationship between two using a function F. E is the error

## Variance  
  - Variance refers to the changes in the result of an algorithm when using different portion/part of the training data.
  - We can define variance as the variability in model prediction when using different datasets.  
  - When we talk about variance affecting the result of an model we usually refer to the result of the **test data not the training data.**  
  - **Low Variance** when our model has low variance it means that our model performs well on the test data and gives good results.  
  - **High Variance** when our model has high variance it means that our model does not perform well on the test data and gives poor results.  
  - Variance of our model will be

### 1. Low Bias and High Variance - OVERFIT
  - When our model has low bias and high variance it means that our model performs well on training dataset and fails to perform on test dataset.  
  - During Low Bias and high variance our model learns too much from the train data and gives good results(accuracy) on train dataset but when an unseen data(test data) is used to predict our model fails to perform and gives poor results(accuracy).  
  
### 2. High Bias and Low Variance - UNDERFIT
  - When our model has high bias and low variance it means our model does not perform well on training data but it gives likely good results(accuracy) on the test data alone. 
  - Mostly during UNDERFIT the result or performance of both the training and testing data will be poor.  
  - A scenario where a machine learning model can neither learn the relationship between variables in the training data nor predict or classify a new data point is called Underfitting.  
  - During such situation the predictions are consistent but the results are inaccurate on average.  
  - It happens when the model does not learn well or only a few number of records are used to train or when used only a few number of parameters.  
    
### 3. High Bias and High Variance - POOR MODEL
  - When our model has high bias and high variance it means that the model does not perform well on both training data and testing data.  
  - Our machine learning model is poor in such situations.  
   
### 4. Low Bias and Low Variance - GOODFIT 
  - When our model has low bias and low variance it means that the model performs well on both training data and testing data.
  - For a model to be effective and accurate the bias and variance has to be low.

# Bias - Variance TradeOff  
  - Our main motive is to build a model with good performance and minimum error to achive that our model has to be a goodfit meaning there should be low bias as well as low variance.  
  - Bias and variance has an inverse relationship.
  - If we reduce the bias it will increase the variance and if we reduce the variance it will increase the bias. _Such a pain right?_  
  - But there is a way to solve this which we just need to find a **balance between bias and variance.**  
  - In other words we need to make sure that our model doesn't learn too much (overfit) and doesn't learn veryless (underfit).
  - The balance between bias-variance is known as bias and variance tradeoff.  
       
  - As shown in the above figure, there exists a point where the error starts going upwards because of the variance increasing and bias decreasing. This is the exact point where the model needs to stop increasing its complexity and use all the parameters defined by that point in the curve.   
  - Usually, this is where the Bias and the Variance curves intersect creating the optimal model complexity point. At this point, the model has low bias and low variance without leading to underfitting or overfitting the model.

## 3.KFold Cross Validation     
  - Kfold cross validation and cross validation is similar in its process.  
  - In Kfold cv we split the data into k folds lets say k=5 we split the data into 5 folds or sets and in each 5 sets we will have train and test data seperately.   
  - The test data will be selected by dividing the total no of records of data with the k value.   
  - For example, we have 1000 records and our k values is 5, the test value will be calculated by 1000/5 which is 200, so out of 1000 records our train data will contain 800 records and test data will contain 200 records. And as our k is 5, we totally have 5 folds or sets of train and test.  
  - The value of K can be between 5 to 10.  
  - The accuracy of 5 different sets can be averaged and the final accuracy will be used to validate our model.  
  - Kfold cross validation can its disadvantage suppose out of 5 sets in our 1st set we only have one type of record in train set i.e there is an imbalance in our train and test data. This will affect the accuracy of our model.

# Outlier  
 - Outlier is a data point which is distant from other data points or observations in an dataset. A data point that lies outside the overall distribution on the data.    
 - Outliers sometimes are very beneficial and sometimes they are not. In a credit card fraud dataset outlier are important as there are only few records that are far away from usual transactions.  
 
## Impacts of outliers  
  1. Affects the statistical analysis  
  2. Causes significant impacts on mean and standard deviation.  
   
## Reasons for outliers  
  1. Variability in the dataset (natural variation).  
  2. Sampling problems.  
  3. Measurement Errors.  
    
## Identifying Outliers  
   1. Box Plot  
   2. Scatter Plot  
   3. Dist plot  


## Handling Outliers
### 1. IQR
   - Inter quartile range is the difference between Q3-Q1. Q3 is 75th percentile and and Q1 is 25th percentile of the data.
   - LOWER BOUND below what value from Q1 a data point is considered as an outlier.
   - UPPER BOUND above what value from Q3 a data point is considered as an outlier.  
   - IQR is used assuming that the dataset follows a normal distribution.  
   <font color=red>sorted(data)  
    Q1,Q3=np.percentile(data,[25,75])  
    IQR=Q3-Q1  
    lower_bound= Q1-(1.5 * IQR)</font>     # 1.5 is most suitable value based on distribution pattern and depends on domain
    <font color=red>upper_bound= Q3+(1.5 * IQR)</font>
   - Inorder to find the extreme outlier value we multiply IQR by 3 std. And 3 std is used especially when the data is skewed.
   
   
   - **Handling Outliers** Once we have found the outliers we can remove it or we can convert those outliers into a normal value. For example lets say age values above 80 are outliers, we have 5 records of 90 as age, we can remove those records or we can convert those 5 records into a lesser age value like 80 as it is the limit.
